```python
import pandas as pd
```

```python
df2017 = pd.read_csv('~/Downloads/happiness_2017.csv')
df2017.head(1)
```

```python
df2012 = pd.read_csv('~/Downloads/happiness_2012.csv')
```

```python
merged = df2017.merge(df2012, suffixes=('_2017', '_2012'), left_on='Country', right_on='country', how='left')
```

```python
merged['Rank_diff'] = merged['Rank_2017']-merged['Rank_2012']
```

```python
# Q-1: What is the name of the country with the largest positive change in their happiness rank?  What about the largest negative change?
merged[['Country', 'Rank_diff', 'HapinessScore_diff']].sort_values('Rank_diff', ascending=True).head(1)
```

```python
merged[['Country', 'Rank_diff', 'HapinessScore_diff']].sort_values('Rank_diff', ascending=False).head(1)
```

```python
merged['HapinessScore_diff'] = merged['HappinessScore']-merged['Happiness Score']
```

```python
# Q-2: Are the countries that with the largest change in rank the same as the countries with the largest change in score?
# Q-3: What is the name of the country with the largest positive change in their happiness score? 
merged[['Country', 'Rank_diff', 'HapinessScore_diff']].sort_values('HapinessScore_diff', ascending=True).head(3)
```

```python
merged[['Country', 'Rank_diff', 'HapinessScore_diff']].sort_values('HapinessScore_diff', ascending=False).head(3)
```

```python
diff_col_names = []
for col_name in ['Life Ladder', 'Log GDP per capita', 'Social support', 'Generosity', 
                 'Negative affect', 'Confidence in national government', 'Positive affect', 'Perceptions of corruption',
                'Freedom to make life choices', 'Healthy life expectancy at birth']:
    new_col_name = col_name.split()[0] + "_d"
    diff_col_names.append(new_col_name)
    merged[new_col_name] = df2017[col_name] - df2012[col_name]
```

```python
# or the five countries with the largest changes changes in ranking between 2012 and 2017 what are the factors that changed the most?
# # Q-5: What did you learn in the previous investigation? What were the factors that changed the most?
col_names_filter=['Country', 'Rank_diff', 'HapinessScore_diff'] + diff_col_names
merged_btm_5_rank = merged[col_names_filter].sort_values('HapinessScore_diff', ascending=True).head(5)
merged_btm_5_rank
```

```python
merged_top_5_rank = merged[col_names_filter].sort_values('HapinessScore_diff', ascending=False).head(5)
merged_top_5_rank
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
lm_expr = 'Rank_diff ~ Life_d + Log_d + Social_d + Generosity_d + Negative_d + Confidence_d + Positive_d + Perceptions_d + Freedom_d + Healthy_d'
mod = ols(lm_expr,
                data=merged).fit()
                
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
# TLDR: nothing stands out in general
```

```python
mod = ols(lm_expr, data=merged_btm_5_rank).fit()
mod.summary()
```

```python
mod = ols(lm_expr, data=merged_top_5_rank).fit()
mod.summary()
```
