```python
# Q-3: The sum of all happiness in South Asia is __ and the sum of all happiness in Western Europe is __ (out to three decimal places)
regions_of_interest = df['Region'].isin(['Western Europe', 'South Asia'])
columns_of_interest = ['HappinessScore', 'Region'] 
df[columns_of_interest][regions_of_interest].groupby('Region').sum()
```

```python
def patch_country(df_target, old_name, new_name):
    df_target.loc[df_target['Country'] == old_name, 'Country'] = new_name
    
patch_country(df_world_countries, 'United States of America',                             'United States')
patch_country(df_world_countries, 'United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')
patch_country(df_world_countries, 'Bosnia & Herzegovina',                               'Bosnia and Herzegovina')
patch_country(df_world_countries, 'Korea, South ', 'South Korea')
patch_country(df_world_countries, 'Korea, North ', 'North Korea')
patch_country(df_world_countries, 'Hong Kong', 'Hong Kong S.A.R., China')
patch_country(df_world_countries, "Côte d'Ivoire", 'Ivory Coast')
patch_country(df_world_countries, 'Taiwan', "Taiwan Province of China")
patch_country(df_world_countries, 'Trinidad & Tobago', 'Trinidad and Tobago')
```

```python
# Q-14: The least happy country in Southeast Asia is
pvt = pd.pivot_table(df_merged, values='HappinessScore', index='Region_left', aggfunc='min')
pvt
```

```python
min_happiness = pvt.unstack().unstack().loc['HappinessScore', 'Southeast Asia']
min_happiness
```

```python
df_merged[df_merged['HappinessScore'] == min_happiness]
```

```python
# Q-15: the most happy country in Southeast Asia is
pvt = pd.pivot_table(df_merged, values='HappinessScore', index='Region_left', aggfunc='max')
max_happiness = pvt.unstack().unstack().loc['HappinessScore', 'Southeast Asia']
max_happiness
df_merged[df_merged['HappinessScore'] == max_happiness]
```
