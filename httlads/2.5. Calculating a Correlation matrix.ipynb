```python
import pandas as pd
```

```python
df2017 = pd.read_csv('~/Downloads/happiness_2017.csv')
```

```python
df_corr = df2017.corr()
df_corr
```

```python
# Q-1: What is the correlation between Freedom to Make life choices and Generosity?
df_corr.loc['Freedom to make life choices','Generosity']
```

```python
# Q-2: Which two factors have the largest positive correlation (not including the diagonal)
col_names = list(df_corr.index.values)
max_so_far = 0;
max_rel = '';
for col_name in col_names:
    df_corr.loc[col_name, col_name] = 0
    this_max_idx = df_corr[col_name].idxmax()
    this_max = df_corr.loc[col_name, this_max_idx]
    if max_so_far < this_max:
        max_rel = col_name + " + " + this_max_idx
        max_so_far = this_max
    df_corr.loc[col_name, col_name] = 1
max_so_far, max_rel
```

```python
# Q-3: Which two factors have the largest negative correlation?
max_so_far = -10;
max_rel = '';
for col_name in col_names:
    df_corr.loc[col_name, col_name] = 0
    col = df_corr[col_name]
    this_max_idx = col[col < 0].idxmax()
    this_max = df_corr.loc[col_name, this_max_idx]

    if max_so_far < this_max:
        max_rel = col_name + " + " + this_max_idx
        max_so_far = this_max
    df_corr.loc[col_name, col_name] = 1
max_so_far, max_rel
```
