```python
#Q-5: average happiness score
round(df['HappinessScore'].mean(), 3)
```

```python
#Q-6: What is the standard deviation of the happiness score?
round(df['HappinessScore'].std(),3)
```

```python
# Q-7: The mean value for healthy life expectancy is
round(df['Healthy life expectancy at birth'].mean(),3), round(df['Healthy life expectancy at birth'].std(),3)
```

```python
# Q-8: What is the maximum value in the generosity column?
df.iloc[df['Generosity'].idxmax()][['Country', 'Generosity']]
```

```python
# Q-11: The country with the lowest generosity score is
df.iloc[df['Generosity'].idxmin()][['Country', 'Generosity']]
```

```python
# Q-12: What is the name of the country that has the highest confidence in their national government?
col_name = 'Confidence in national government'
df.iloc[df[col_name].idxmax()][['Country', col_name]]
```

```python
# Q-13: What is the name __ and happiness score __ of the country with the lowest confidence in their national government?
col_name = 'Confidence in national government'
df.iloc[df[col_name].idxmin()][['Country', col_name, 'HappinessScore']]
```

```python
df.corr().loc['HappinessScore']
```

```python
# correlations for top 25 countries
top_corr = df.sort_values('HappinessScore', ascending=False).head(25).corr().loc['HappinessScore']
top_corr
```

```python
# correlations for bottom 25 countries
bottom_corr = df.sort_values('HappinessScore', ascending=True).head(25).corr().loc['HappinessScore']
bottom_corr
```

```python
# 5. We might now try to focus in on the characteristics of the most happy countries and the least happy countries.
(top_corr - bottom_corr).sort_values(ascending=False)
```

```python
sort_col = 'Social support'
val_col = 'HappinessScore'
sorted_df = df.sort_values(sort_col, ascending=True)
sorted_df[f'{val_col}_above'] = sorted_df[val_col].shift(-1)
sorted_df[f'{val_col}_below'] = sorted_df[val_col].shift(1)
sorted_df[f'{val_col}_diff'] = (sorted_df[f'{val_col}_above']-sorted_df[f'{val_col}']).abs() + \
                               (sorted_df[f'{val_col}_below']-sorted_df[f'{val_col}']).abs()
sorted_df.sort_values(f'{val_col}_diff', ascending=False)
```
